## 3- Data cleaning & Structuring

```python
# the exact localisation has a function that contains the target position
# this columns is in location,vente and ImmoNeuf tables
```

```python
## Clean Immobilier Neuf
```

```python
def extract_text(string, start_str, end_str):
    """
    Extracts the text from a string between the start_str and end_str substrings.
    Returns an empty string if either substring is not found in the string.
    """
    start_index = string.find(start_str)
    if start_index == -1:
        return ""
    end_index = string.find(end_str, start_index + len(start_str))
    if end_index == -1:
        return ""
    return string[start_index:end_index]
```

```python
Immo_neuf["Exact_localisation"]=Immo_neuf["Exact_localisation"].apply(extract_text,args=("http","'"))
vente["exact_local"]=vente["exact_local"].apply(extract_text,args=("http","'"))
location["exact_local"]=location["exact_local"].apply(extract_text,args=("http","'"))
```

```python
def clean_description(l_text):
    string=""
    for text in l_text:
        text = text.replace('\n', ' ').replace('\t', ' ').replace('\r', ' ')
        string+=text
    return string
```

```python
Immo_neuf['Description']=Immo_neuf['Description'].apply(clean_description)
```

```python
def clean_price(text):
    text=text.replace("DH",'').replace("À partir de","")
    return text
```

```python
Immo_neuf['price']=Immo_neuf['price'].apply(clean_price)
Immo_neuf.rename(columns={'price': 'price (À partir de)'}, inplace=True)
```

```python
## Clean Vente
```

```python
vente['price']=vente['price'].apply(clean_price)
vente.rename(columns={'price': 'price (DH)'}, inplace=True)
```

```python
def clean_tags(l_text):
    string=[]
    for text in l_text:
        text = text.replace('\n', '').replace('\t', '').replace('\r', '')
        string.append(text)
    return string
```

```python
vente['tags']=vente['tags'].apply(clean_tags)
```

```python
vente["Description"]=vente["Description"].apply(clean_description)
```

```python
vente["Caracteristiques"]=vente["Caracteristiques"].apply(clean_tags)
```

```python
def clean_agenc_name(text):
    try:
        text=text.replace("\n","").replace("\t","").replace("\r","")
        return text
    except:
        pass
```

```python
vente["agence_name"]=vente["agence_name"].apply(clean_agenc_name)
```

```python
## clean location
```

```python
location['price']=location['price'].apply(clean_price)
location.rename(columns={'price': 'price (DH)'}, inplace=True)
```

```python
location['tags']=location['tags'].apply(clean_tags)
location['Description']=location['Description'].apply(clean_description)
location["Caracteristiques"]=location["Caracteristiques"].apply(clean_tags)
```

```python
## Clean locationVac
```

```python
locationVac['price']=locationVac['price'].apply(clean_price)
locationVac.rename(columns={'price': 'price (DH)'}, inplace=True)
```

```python
locationVac["tags"]=locationVac["tags"].apply(clean_tags)
locationVac["Description"]=locationVac["Description"].apply(clean_description)
locationVac["Caracteristiques"]=locationVac["Caracteristiques"].apply(clean_tags)
locationVac["exact_local"]=locationVac["exact_local"].apply(extract_text,args=("http","'"))
```

```python
## Clean Agencies
```

```python
agencies.head()
```

```python
agencies["phone_number"]=agencies["phone_number"].apply(lambda x: x.replace('tel:',""))
```

```python
def extract_website(l):
    if len(l)>1:
        if 'http' in l[1]:
            return l[1]
        else:
            return ''
    else:
        return ''
```

```python
agencies["website_if_exists"]=agencies["website_if_exists"].apply(extract_website)
```

```python
agencies['number_of_properties']=agencies['number_of_properties'].apply(lambda x : x.split()[2].replace("résultats",""))
```

```python
agencies["property_title"]=agencies["property_title"].apply(clean_tags)
```

```python
agencies["property_price"]=agencies["property_price"].apply(clean_tags)
```

```python
def clean_prices(l):
    s=[]
    for price in l:
        s.append(price.strip().replace("\xa0",""))
    return s
```

```python
agencies["property_price"]=agencies["property_price"].apply(clean_prices)
```
